```python
pop_Seoul.rename(columns={pop_Seoul.columns[0] : '구별',
                          pop_Seoul.columns[1] : '인구수',
                          pop_Seoul.columns[2] : '한국인',
                          pop_Seoul.columns[3] : '외국인',
                          pop_Seoul.columns[4] : '고령자',
                            }, inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul.drop([0], inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul['외국인비율'] = pop_Seoul['외국인']/pop_Seoul['인구수']*100
pop_Seoul.head()
```

```python
pop_Seoul['고령자비율'] = pop_Seoul['고령자']/pop_Seoul['인구수']*100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by='고령자비율', ascending=False).head(5)
```

```python
CCTV_seoul['최근증가율'] = (CCTV_seoul['2016년'] + 
                           CCTV_seoul['2015년']+
                           CCTV_seoul['2014년']) / CCTV_seoul['2013년도 이전'] * 100
CCTV_seoul.head()
```

```python
CCTV_seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

```python
data_result =  pd.merge(CCTV_seoul, pop_Seoul, on = '구별')
data_result.head()
```

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']

data_result.head()
```

```python
data_result.set_index('구별', inplace=True) #기존의 index사라짐.(대체)
```

```python
data_result.head()
```

```python
import numpy as np
```

```python
# CCTV수와 상관있을 것 같은 항목을 상관분석으로 찾아서,
# 그 항목으로 결정을 해보려고 함.
```

```python
# 고령자비율, 외국인이율, 인구수 중 어떤 것이 CCTV소계와 상관이 있는지 찾아보자.!!!
```

```python
np.corrcoef(data_result['고령자비율'], data_result['소계']) #상관분석=> 고령자비율과 CCTV소계와는 상관없이 없다. 있다면 음의 상관!
```

```python
# 10%이내의 데이터 값은 거의 의미가 없는 수치임.
# 보통 30%이하의  데이터 값은 크게 의미가 없는 수치임.
# 30%이상 70%이하였을 때 의미있는 데이터라고 여김.
```

```python
np.corrcoef(data_result['외국인비율'], data_result['소계']) #상관분석=> 외국인비율과 CCTV소계와는 상관없이 없다. 있다면 음의 상관!
```

```python
np.corrcoef(data_result['인구수'], data_result['소계']) 
```

```python
# #상관분석=> 인구수와 CCTV소계와는 크게 상관없이 없다. 하지만, 다른 항목보다 높은 수치를 나타내는 것은 사실임.!
```

```python
t = np.arange(0, 12, 0.01)
y = np.sin(t)
z = np.cos(t)

plt.figure(figsize=(10,6))

plt.plot(t, y, label='sin', lw=5)
plt.plot(t, z, 'g^', label='cos')
plt.legend()

plt.grid()
plt.xlabel('time')
plt.ylabel('yvalue')
plt.title('my graph')

plt.show()
```

```python
plt.figure(figsize=(10,6))

t = np.arange(0, 5, 0.5)

plt.plot(t, t, 'r--')
plt.plot(t, t*2, 'bs')
plt.plot(t, t*3, 'g^')
plt.plot(t, t*4, color='green', linestyle='dashed', marker='o')



plt.show()
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)

plt.figure(figsize=(10,6))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot((s1, s2, s3))
plt.grid()
plt.show()
```
