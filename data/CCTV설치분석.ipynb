```python
pop_Seoul = pd.read_excel('pop_seoul.xls', encoding='utf-8')
```

```python
pop_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('pop_seoul.xls',
                         header=2,
                         usecols='B, D, G, J, N',
                         encoding='utf-8')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0]:'구별',
                          pop_Seoul.columns[1]:'인구수',
                          pop_Seoul.columns[2]:'한국인',
                          pop_Seoul.columns[3]:'외국인',
                          pop_Seoul.columns[4]:'고령자'
                         }, inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul.head()
```

```python
pop_Seoul[pop_Seoul['구별'].isnull()]
```

```python
pop_Seoul['외국인비율'] = pop_Seoul['외국인']/pop_Seoul['인구수']*100
pop_Seoul.head()
```

```python
pop_Seoul['고령자비율'] = pop_Seoul['고령자']/pop_Seoul['인구수']*100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by='인구수', ascending=True)
```

```python
pop_Seoul.sort_values(by='외국인비율', ascending=False)
```

```python
CCTV_seoul['최근증가율'] = (CCTV_seoul['2016년'] +
                       CCTV_seoul['2015년'] +
                       CCTV_seoul['2014년']) / CCTV_seoul['2013년도 이전'] * 100
CCTV_seoul.head()
```

```python
CCTV_seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

```python
CCTV_seoul.sort_values(by='2016년', ascending=False).head(5)
```

```python
t = np.arange(0,12,0.01)
y = np.sin(t)
z = np.cos(t)
```

```python
plt.figure(figsize=(10,6))
plt.plot(t,y, label='sin', lw=5)
plt.plot(t,z, 'g^', label='cos')
plt.legend()
plt.grid()
plt.xlabel('time')
plt.ylabel('yvalue')
plt.title('my graph')
plt.show()
```

```python
plt.figure(figsize=(10,6))

t = np.arange(0,5,0.5)

plt.plot(t,t,'r--')
plt.plot(t,t*2, 'bs')
plt.plot(t,t*3, 'g^')
plt.plot(t,t*4, color='green', linestyle='dashed', marker='o')
plt.show()
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)

plt.figure(figsize=(10,6))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot((s1, s2, s3))
plt.grid()
plt.show()
```
